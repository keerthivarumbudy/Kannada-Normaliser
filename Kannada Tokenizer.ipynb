```python
import pickle
import pathlib
```

```python
file = './KannadaWikipediaArticles/10500.pkl'
```

```python
f = open(file, 'rb')
text = pickle.load(f)
#Splitting article into paragraphs and indexing them in a dictionary
list_paragraphs = text.split('\n')
list_paragraphs = list(filter(lambda a: a!='',list_paragraphs))
map_paragraphs = dict()
for i in range(1,len(list_paragraphs)+1):
    map_paragraphs[i] = list_paragraphs[i-1]
map_paragraphs
```

```python
'''for i in map_paragraphs:
    list_sentences.appendmap_paragraphs[i].split('.')'''
temp_list_sentences = [x.split('.') for x in map_paragraphs.values()]
temp_list_sentences
```

```python
#print(len(list_paragraphs))
list_sentences = [[] for i in range(len(list_paragraphs))]
counter = 0
for i in range(len(list_paragraphs)):
    for j in range(len(temp_list_sentences[i])):
        #print(list_sentences, counter)
        if counter!=0:
            #print(i,j)
            list_sentences[i][-1] = list_sentences[i][-1] + '.' + temp_list_sentences[i][j]
            counter = 0
            pass
        
        elif len(temp_list_sentences[i][j])>3:
            #print(list_sentences)
            list_sentences[i].append(temp_list_sentences[i][j])
            
        elif temp_list_sentences[i][j]=='':
                pass
            
        else:
            list_sentences[i][-1] = list_sentences[i][-1] + '.' + temp_list_sentences[i][j]
            counter = 1
            
list_sentences
            
```

```python
map_sentences = dict()
for i in range(1,len(list_paragraphs)+1):
    sentence_index = dict()
    for j in range(1, len(list_sentences[i-1])+1):
        if list_sentences[i-1][j-1] != '':
            sentence_index[j] = list_sentences[i-1][j-1]
    map_sentences[i] = sentence_index
map_sentences
```

```python
map_tokens = map_sentences.copy()
tokens = dict()
for i in range(1,len(map_paragraphs)+1):
        for j in range(1,len(map_sentences[i])+1):
            sentence = map_sentences[i][j].split('| , | ; ')
            sentence = str(sentence).split(' ')
            for k in range(1,len(sentence)+1):
                tokens[k] = sentence[k-1]
            map_tokens[i][j]=tokens
print(map_tokens)
```
